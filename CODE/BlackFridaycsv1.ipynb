```
import matplotlib.pyplot as plt
 
# Data to plot
labels = 'M', 'F'
sizes = [405380,132197]
colors = ['blue', 'red']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.show()
```
